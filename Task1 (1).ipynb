```python
import time
from datetime import datetime
import pandas as pd
#convert it to the format of last modified
today = time.strftime("%Y-%m-%d")
#convert it to string
today = str(today)
from usp.tree import sitemap_tree_for_homepage
tree = sitemap_tree_for_homepage('https://www.exportleftovers.com/')
for page in tree.all_pages():
  try:
    pagdate = str(page.last_modified)
    date=pd.to_datetime(pagdate).date()
    if str(today) == str(date): 
      print(page.url , page.last_modified)  
      with open ("/content/drive/MyDrive/Cognilium/datewiseelo.txt" , 'a') as f:
        f.write(str(page.url)+","+str(page.last_modified)+"\n")
        f.close()
  except:
    pass 
print(today,date)


# tree = sitemap_tree_for_homepage("https://brandsego.com/")
# # https://brandsego.com/
# #https://brandspopper.com/
# print(tree)
```
